# Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("Zomato data .csv")
```

```python
print(dataframe)
```

```python
dataframe
```

# here we are removing /5 from the ratings(Data cleaning )

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# finding the type of res from where the people are ordering


```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("types of restuarant")
```

# Conclusion - Dining

# Votes gained by the different type of restaurant

```python
group_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result= pd.DataFrame({'votes': group_data})
plt.plot(result,c="red",marker="o")
plt.xlabel("types of restuarant",c="blue")
plt.ylabel("no of votes",c="blue")
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("rating distribution")
plt.show()
```

# conclusion:- Majority rating are between 3.5 to 4

# Avearge order spent by couples 

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
sns.set_style('whitegrid')
```

# Conclusion:- the majority of  approx cost for two people is Rs.300/-

# which mode receives more rating

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order' , y= 'rate' , data= dataframe)
```

# Conclusion :- Maximum rating is being received by online orders

# type of resturant that received more offline orders

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)', columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("heatmap")
plt.xlabel("online orders")
plt.ylabel("listed in (Type)")
plt.show()
```

# Conclusion:- Dining received more offline orders whereas cafes are receiving more online orders
